## 现在来介绍pandas的map函数

 通过一个字典将一列的值进行映射

```python
mydict={
    '1':"无学术背景",
    '2':"无学术背景",
    '3':"无学术背景",
    '4':"有学术背景",
}

print(data['学术背景'].map(mydict))
```

当然这里也可以是一个函数，甚至是一个列（Series），简化版的apply
map使用的遍历所有行的方式，对每个值进行映射，map函数只能应用在一个列上

```python
#传入函数
def mapfunc(x):
    #这里的x值得是每个行的值
    #这里就可以对其进行任意的操作了
    #比如
    if x=="1":
        return (int(x)+1)*3
    elif x=="2":
        return "aaaa"
    else:
        return "bbbb"        

print(data['学术背景'].map(mapfunc))
```

这里的聚合是可以传入函数的,使用apply()即可

```python
grouped1.apply(len)
```

apply可以用做列也可以用作整个数据框
在对一个Series使用apply的时候
相当于遍历这个列的每个值，并对每个值应用函数
例如

```python
def app_series(x):
    #这里的x代表 列中的每一值
    #这里我们处理年份将所有日期年份都只取其年不要月和日
    return x.split('-')[0]
data['统计截止日期'].apply(app_series)
```

假如是对整个数据框呢？apply传入的应该是什么

```python
def app_dataframe(col):
    #这里的col是列，即每次默认传入的是这个数据框的列
    #当然也可以设置传入行
    #下面是axis设置为1，传入行的代码
    col['是否高管团队成员'] = '是' if col['是否高管团队成员']== 1 else '否'
    col=col[['证券代码','统计截止日期','人员ID','是否高管团队成员','学术背景']]

    return col
data.apply(app_dataframe,axis=1)   
```

apply可以和分租很好的配合。
因为groupby函数之后的group对象，对他进行迭代，每一次都拿到的是一个数据框
所以可以做如下操作

```python
def app_func(df):
    return (1-len(df[df['学术背景']=='4'])/len(df))

grouped1.apply(app_func)
```

计算当年中每个公司的学术高管占比比列

apply还可以传入自定义的参数

现在我们运用一个完整的例子来正式进行一波pandas的分析，
现在我要求统计所有的上市公司当中，高管有学术背景的占比

```python
data=pd.read_excel('高.xlsx')
data=data[['证券代码','统计截止日期','人员ID','是否高管团队成员','学术背景']]
data
```

```python
#处理日期
def date_map(x):
    return x.split('-')[0]
data['统计截止日期']=data['统计截止日期'].map(date_map)
data['统计截止日期']
```

```python
#剔除不是高管的
data=data[data['是否高管团队成员']==1]
data
```

```python
grouped=data.groupby(['统计截止日期','证券代码'])
print(grouped.size())
#现在groupname是一个元组
```

```python
#统计学术高管占每个每组的比值
def app_func(df):
    return (1-len(df[df['学术背景']=='4'])/len(df))

out=grouped.apply(app_func)
out.to_excel('占比结果.xlsx')
```

```python
#现在out拥有多级索引
for index,data in out.items():
    print(index)
```
